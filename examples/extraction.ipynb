```python
import torch
from extraction.extractor import TrafficDataExtractor
from extraction.collector import TrafficDataCollector
from commonroad_geometric_io.dataset.extraction.traffic.edge_drawers import FullyConnectedEdgeDrawer
import random
```

# Extract Graph Data

```python
def collect_traffic_data(input_dir, output_dir):
    input_directory = input_dir
    collector = TrafficDataCollector(
        extractor_cls=TrafficDataExtractor,
        extractor_kwargs=dict(
            assume_straight_lanelets=True,
            edge_drawer=FullyConnectedEdgeDrawer()
        )
    )
    collector.export_from_directory(
        input_dir=input_directory,
        output_dir=output_dir,
        max_samples_per_scenario=100,
        max_total_scenarios=1000,
        overwrite=True,
        shuffle_scenarios = False,
        skip_subvariants = False,
        dataset_size_percent=(80, 20)
    )

def show_random_sample(output_dir, name):
    data = torch.load(f'{output_dir}/{name}.pth')
    if len(data) > 0:
        random_idx = random.randrange(len(data))
        print(random_idx, data[random_idx])
    else:
        print("No samples found in", name)


collect_traffic_data("test_data", "output")
show_random_sample("output", "training")
show_random_sample("output", "test")
```
